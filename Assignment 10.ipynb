# 1. How do you distinguish between shutil.copy() and shutil.copytree()?

Ans:- 
    
    While shutil.copy() will copy a single file,
    
    shutil.copytree() will copy an entire folder and every folder and file contained in it.

# 5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.

```python
import os, shutil
```

```python
def movefile(folders):
    for r, d, k in os.walk(folders):
        for f in k:
            if f.endswith('.pdf'):
                pdf_path = os.path.join(r,f)
                print('location: ', pdf_path)
                shutil.copy(pdf_path, 'C:/Users/Admin/Ineuron/Ineuron Assignment/Transfer')
                
#r --> root
#d --> dirs
#k --> files
```
