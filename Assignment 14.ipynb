# 1. What does RGBA stand for?

Ans:-
    
    The RGBA color format is an extension of the RGB scheme with an added alpha channel that specifies the opacityt of the 
    color.
    The RGBA for fornat can be specified using:
    
    rgba(red, green, blue, alpha)
    
    The alpha value is declared as decimal number from 0 to 1. where 0 us fully transparent and 1 is fully opaque.

```python
 from PIL import Image
```

```python
img = Image.open('img1.png')
rgba =  img.convert('RGBA')
dataa = rgba.getdata()
```

```python
newdata = []
for item in dataa:
    if item[0] ==0 and item[1] == 0 and item[2] == 0:
        newdata.append((255,255,255,0))
    else:
        newdata.append(item)

rgba.putdata(newdata)
rgba.save('tranparent_image.png', 'PNG')
```
