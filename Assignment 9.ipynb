# 6. What are the three “mode” arguments that can be passed to the open() function?

Ans:-
    
    The "modes" arguments that can be passed to the open function are as follows:
        
    1) 'w' -  Open a file for writing mode. Creates a new file if it does not exist or truncates the file if it exists.
    
    2) 'r' -  Open a file for reading.
    
    3) 'a' -  Open for appending at the end of the file without truncating it. Creates a new file if it does not exist.

# 7. What happens if an existing file is opened in write mode?

Ans:
    
    If an existing file is opened in write mode. its contents are discarded and the file is treated as a new empty file.
